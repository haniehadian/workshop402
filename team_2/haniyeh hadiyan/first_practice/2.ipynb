```python
import random


counter = 1

print(f'"Celsius"\t\t\t\t\t\t  "Fahrenheit"')

while counter <= 10 :
    
    Temperature = round( random.random() * (100 - 0) + 0 , 5 )
    
    print(f'\n{counter}) {Temperature}°C \t  { ((9/5) *Temperature) + 32:50}°F')
    
    print('-'*80)
    
    counter += 1
```

```python
binary = []

counter = 1

print('\n--> Enter your 9-bit binary string')

while counter <= 9 :
    
    binary.append( int( input(f'{counter} bit is :') ) )  
    
    counter += 1          
#..

if sum(binary) % 2 == 0 :
    
    print(f'Binary string is even parity')

else :
    
    print('-'*40)
    print(f'--> Binary string is not even parity')
```

```python
x = float(input('Please enter your number : '))

guess = x / 2

while abs(guess * guess - x) < 0.001 :
    
    guess = (guess + (x / guess)) / 2

print('Final guess is :',guess)
```
